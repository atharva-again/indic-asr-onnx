```python
# Run the eval

import tempfile

import numpy as np
import torch
import torchaudio
from datasets import load_dataset
from universal_edit_distance import character_error_rate, word_error_rate

from indic_asr_onnx import IndicTranscriber


def get_audio_path(sample):
    audio_path = sample["audio_filepath"]
    waveform = None
    sample_rate = 16000

    if hasattr(audio_path, "get_all_samples"):  # TorchCodec AudioDecoder
        decoded = audio_path.get_all_samples()
        waveform = decoded.data.float()
        sample_rate = decoded.sample_rate
    elif isinstance(audio_path, dict) and "array" in audio_path:  # Decoded dict
        waveform = torch.from_numpy(audio_path["array"]).float()
        sample_rate = audio_path["sampling_rate"]
    elif isinstance(audio_path, np.ndarray):  # Numpy array
        waveform = torch.from_numpy(audio_path).float()

    # If we extracted a waveform, save it to a temp file
    if waveform is not None:
        if waveform.ndim == 1:
            waveform = waveform.unsqueeze(0)  # (1, T)
        with tempfile.NamedTemporaryFile(suffix=".wav", delete=False) as tmp_file:
            torchaudio.save(tmp_file.name, waveform, sample_rate)
            audio_path = tmp_file.name
    elif not isinstance(audio_path, str):
        raise ValueError(f"Could not extract audio from {type(audio_path)}")
    return audio_path


def main():
    # Load the quantized model
    transcriber = IndicTranscriber()

    # Load the Hindi validation set
    dataset = load_dataset(
        "ai4bharat/IndicVoices", "hindi", split="valid", streaming=True
    )

    ctc_predictions = []
    rnnt_predictions = []
    references = []
    counter = 0

    print("Processing samples...")

    for sample in dataset:
        audio_path = get_audio_path(sample)
        ground_truth = sample["normalized"]
        references.append(ground_truth)

        counter += 1
        if counter % 100 == 0:
            print(f"Processed {counter} samples...")

        ctc_pred = transcriber.transcribe_ctc(audio_path, "hi")
        ctc_predictions.append(ctc_pred)

        rnnt_pred = transcriber.transcribe_rnnt(audio_path, "hi")
        rnnt_predictions.append(rnnt_pred)

    num_samples = len(references)
    print(f"Processed {num_samples} samples from the Hindi validation set.")

    # Compute mean WER and CER for CTC
    wer_ctc = word_error_rate(ctc_predictions, references)
    cer_ctc = character_error_rate(ctc_predictions, references)
    print(f"CTC - Mean WER: {wer_ctc:.4f}")
    print(f"CTC - Mean CER: {cer_ctc:.4f}")

    # Compute mean WER and CER for RNNT
    wer_rnnt = word_error_rate(rnnt_predictions, references)
    cer_rnnt = character_error_rate(rnnt_predictions, references)
    print(f"RNNT - Mean WER: {wer_rnnt:.4f}")
    print(f"RNNT - Mean CER: {cer_rnnt:.4f}")


if __name__ == "__main__":
    main()
```
